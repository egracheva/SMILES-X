```python
import pandas as pd

from SMILESX import embedvisual, interpret, token
%matplotlib inline
```

#### Definition of data

```python
validation_data_dir = "./data/"
```

```python
extension = '.csv'
```

```python
data_name = 'FreeSolv' # FreeSolv, ESOL, Lipophilicity
prop_tag = ''
```

```python
if data_name == 'FreeSolv':
    data_filename = 'FreeSolv_SAMPL'
    prop_tag = 'expt'
elif data_name == 'ESOL':
    data_filename = 'ESOL_delaney-processed'
    prop_tag = 'measured log solubility in mols per litre'
elif data_name == 'Lipophilicity':
    data_filename = 'Lipophilicity'
    prop_tag = 'exp'
else:
    data_filename = data_name
    prop_tag = prop_tag
```

```python
sol_data = pd.read_csv(validation_data_dir+data_filename+extension)
```

```python
sol_data.head(3)
```

```python
sol_data = sol_data[['smiles',prop_tag]]
```

```python
sol_data.head()
```

```python
sol_data.shape
```

#### Embedding Visualization

```python
embedvisual.Embedding_Vis(data=sol_data, 
                            data_name=data_name, 
                            data_units='', 
                            k_fold_number = 2,
                            k_fold_index = 0,
                            augmentation = True, 
                            outdir = "./data/", 
                            affinity_propn = True)
```

#### Find tokens in SMILES

```python
token.TokensFinder(data=sol_data, 
                           data_name=data_name, 
                           data_units='', 
                           k_fold_number = 1,
                           k_fold_index = 0,
                           augmentation = True, 
                           token_tofind = '[nH]', 
                           verbose = 0)
```

#### SMILES inference interpretation

```python
interpret.Interpretation(data=sol_data, 
                             data_name=data_name, 
                             data_units='', 
                             k_fold_number = 1,
                             k_fold_index = 0,
                             augmentation = True, 
                             outdir = "./data/", 
                             smiles_toviz = 'ClC1=C(Cl)[C@]2(Cl)[C@H]3[C@H]4C[C@H]([C@@H]5O[C@@H]54)[C@H]3[C@@]1(Cl)C2(Cl)Cl', 
                             font_size = 15, 
                             font_rotation = 'vertical')
```
